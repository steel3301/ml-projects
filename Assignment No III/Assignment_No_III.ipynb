```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('admission.csv')
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.nunique()
```

```python
sns.barplot(x='University Rating', y='CGPA', data=data)
plt.title('University Rating vs Average CGPA')
plt.show()
```

```python
sns.boxplot(x='SOP', y='Chance of Admit', data=data)
plt.title('SOP vs Chance of Admit')
plt.show()
```

```python
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.plot(data['GRE Score'], data['Chance of Admit'], marker='o', linestyle='-')
plt.title('Trend of GRE Score vs Chance of Admit')
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')
plt.show()
```

```python
data['CGPA'].hist(bins=10, color='green')
plt.title('Distribution of CGPA')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```
