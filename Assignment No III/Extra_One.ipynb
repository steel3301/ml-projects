```python
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
actual = np.array(['Spam','Spam','Spam','Spam','Not Spam','Spam','Spam','Not Spam','Spam','Spam'])

predicted = np.array(['Spam','Spam','Spam','Spam','Spam','Spam','Not Spam','Not Spam','Not Spam','Spam'])
```

```python
cm = confusion_matrix(actual,predicted)
```

```python
sns.heatmap(
    cm,
    annot = True,
    fmt = 'g',
    xticklabels = ['Spam','Not Spam'],
    yticklabels = ['Spam','Not Spam']
)

plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Example of handling multi-class
accuracy = accuracy_score(actual, predicted)
print('Accuracy:', accuracy)

precision = precision_score(actual, predicted, average='weighted')  # Specify 'average'
print('Precision:', precision)

recall = recall_score(actual, predicted, average='weighted')  # Specify 'average'
print('Recall:', recall)

f1 = f1_score(actual, predicted, average='weighted')  # Specify 'average'
print('F1 Score:', f1)
```
