```
#e) Finding out Zero's
```

````python
zeros = (data == 0).sum().sum()  # Total count of zeros
print(f"Total zeros in the dataset: {zeros}")

# Find zeros column-wise
column_zeros = (data == 0).sum()
print("Zeros in each column:")
print(column_zeros)

# Find rows with zeros
rows_with_zeros = data[(data == 0).any(axis=1)]
print("Rows containing zeros:")
print(rows_with_zeros)
````

```
#f) Indexing and selecting data, sort data,  
```

````python
df = pd.DataFrame(data)
````

````python
# Selecting specific rows
print(df.iloc[1])  # Select by row index (positional)
print(df.loc[1])   # Select by row label (index)

# Selecting rows and columns
print(df.iloc[1, 2])  # Select specific cell (row 1, column 2)
print(df.loc[1, 'Age'])  # Select by label
````

````python
sorted_df = df.sort_values(by='Age', ascending=True)  # Sort by 'Age' in ascending order
print(sorted_df)
````
